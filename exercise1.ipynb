# Exercise 1

### 1a

disparity: d = x1-x2

x1 / f = px / pz,

-x2 / f = (T - px) / pz

=> x1 = px / pz * f

=> -x2 = f * T / pz - f * px / pz

d = x1-x2 = f * px / pz - f * px / pz + f * T / px = f * T / pz

T and f constant => bigger pz smaller disparity smaller pz bigger disparity

### 1b

```python
def disparity(pz, T, f):
    return T * f / pz

T = 0.0025
f = 0.12
pzArr = np.arange(0.1, 10, 0.1) # start, stop, step
disparitiesArr = disparity(pzArr, T, f)

fig, a = plt.subplots(1, 1)
fig.set_figwidth(10)
fig.set_figheight(10)
a.set_title("Disparity - distance")
a.set_xlabel("asd")
# a.set(xlim=(0, 10), ylim=(0, 0.01))
a.plot(pzArr, disparitiesArr)
plt.show()


# f, a = plt.subplots(1, 4)
# f.tight_layout()
# f.set_figwidth(10)
# f.set_figheight(3)
# a[0].set_title("Original")
# a[0].set(xlim=(0, 40), ylim=(0, 4))
# a[0].plot(temp)
# a[1].set_title("Corrupted")
# a[1].set(xlim=(0, 40), ylim=(0, 4))
# a[1].plot(corrupted)
# a[2].set_title("Gauss")
# a[2].set(xlim=(0, 40), ylim=(0, 4))
# a[2].plot(gaussSignal)
# a[3].set_title("Median")
# a[3].set(xlim=(0, 40), ylim=(0, 4))
# a[3].plot(medianSignal)
```
